```python
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
Z1 = np.random.normal(size=10000)
Z2 = np.random.normal(size=10000)
```

```python
chisq_12 = Z1**2 + Z2**2
```

```python
chisq_df2 = np.random.chisquare(df=2,size=10000)
```

```python
sns.distplot(chisq_12);
sns.distplot(chisq_df2);
```

```python
chisq_1212 = Z1**2 + Z2**2 + ((Z1+Z2)/2)**2
```

```python
chisq_df3 = np.random.chisquare(df=3,size=10000)
sns.distplot(chisq_1212);
sns.distplot(chisq_df3);
```

```python
stats.chi2.isf(0.05,df=30-1)
```

```python
stats.chi2.ppf(0.95,df=30-1)
```

```python
stats.chi2.sf(23.49, df=30-1)
```
