```python
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
Z1 = np.random.normal(size=10000)
Z2 = np.random.normal(size=10000)
Z3 = np.random.normal(size=10000)
Z4 = np.random.normal(size=10000)
```

```python
chisq_jugad_4 = Z1**2 + Z2**2 + Z3**2 + Z4**2
```

```python
chisq_df_4 = np.random.chisquare(df=4,size=10000)
```

```python
sns.distplot(chisq_jugad_4)
sns.distplot(chisq_df_4)
```

```python
np.mean(chisq_df_4), np.var(chisq_df_4)
```

```python
chisq_df_50 = np.random.chisquare(df=50,size=10000)
stats.shapiro(chisq_df_50)
```

```python
sns.distplot(chisq_df_50)
```
